### Verification of NaN data

```python
# Verification of Null data by percent
amazon_df.isnull().sum()*100 / len(amazon_df)
```

```python
def cleanPercentNaN(dataframe: pd.DataFrame, percent:float):
    """Removes variables from the dataset according to the minimum percentage of null values 
    ​​it contains, this given by analyst"""

    percentNaN = (dataframe.isnull().sum() * 100) / len(dataframe)
    list_remove = percentNaN[percentNaN >= percent].index
    list_remove = list(list_remove)
    pd_removed = dataframe.drop(columns = list_remove)

    return pd_removed
```

```python
amazon_df_wn = cleanPercentNaN(amazon_df,30)
amazon_df_wn
```

```python
amazon_df_wn.isnull().sum()*100 / len(amazon_df_wn)
```

- Three of the dataset's features are in their majority NaN values, consequently let's erase those. In fact, they aren't relevant for analysis.

### < 10% NaN values

```python
# values with less of 10% of NaN Values
amazon_df_wn.isnull().sum()*100 / len(amazon_df_wn)
```

```python
# Dataset from list of less 10% NaN values
less_10percent = amazon_df_wn.isnull().sum()*100 / len(amazon_df_wn)
```

```python
# Names of the index of list of less 10% NaN values
index_less_10percent = less_10percent[(less_10percent < 10) & (less_10percent > 0)].index
index_less_10percent
```

```python
# reemplace NaN values with mode value
for i in index_less_10percent:
    # print(amazon_df[i])
    amazon_df_wn[i] = amazon_df_wn[i].fillna(amazon_df_wn[i].mode()[0])
```

```python
# "mode" of values with less of 10% NaN values
for i in index_less_10percent:
    print(amazon_df_wn[i].value_counts(), "La moda es: ", amazon_df[i].mode()[0])
```

```python
# Visualize the new NaN values in percert
amazon_df_wn.isnull().sum()*100 / len(amazon_df_wn)
```

Conclusions:
- Elimination of NaN values very extended, and data imputation of the NaN values less than 10% throught mode value from each variable that requires doing so.

```python
"""
Create a function that describes the type of variables and also statistics regarding the mode, mean, median, 
both in categorical and quantitative variables.
In this function, you can also find out the percentage of nulls, and atypical or repeated values.
"""

import pandas as pd

def variable_statistics(dataframe: pd.DataFrame):
    statistics = {
        'Total_elements': dataframe.notnull().sum(),
        'NaN_amount': dataframe.isnull().sum(),
        'NaN_percent': (dataframe.isnull().sum() * 100) / len(dataframe),
        'Duplicated_data': [dataframe.duplicated().sum()] * len(dataframe.columns),
        'Data_type': dataframe.dtypes,
        'Mode': dataframe.apply(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
    }
    
    pd_statistics = pd.DataFrame(statistics)
    return pd_statistics
```
